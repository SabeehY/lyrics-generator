```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as plticker
```

```python
title = "Pink Floyd - New Model [Dropout 0.2]"
ax = pd.read_csv('pink_floyd_new_model_100.csv', skiprows=3) \
  .set_index('Epoch') \
  .plot(title=title, figsize=(15, 8))

loc = plticker.MultipleLocator(base=5.0) # this locator puts ticks at regular intervals
ax.xaxis.set_major_locator(loc)
```
